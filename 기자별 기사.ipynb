```python
import requests
import lxml.html
import re
```

```python
url = 'http://news.kbiz.or.kr/news/articleList.html?page={}&total=36948&box_idxno=&sc_section_code=S1N1&view_type=sm'#페이지 비우기
```

```python
page = 1
res = requests.get(url.format(page))
```

```python
res
```

```python
root = lxml.html.fromstring(res.text)
```

```python
writers = root.cssselect('div.list-dated')
```

```python
writers
```

```python
content = writers[0].text_content()
```

```python
content = re.sub(r'\d', ' ', content)
```

```python
content = re.findall('\S+기자\S+',content)
```

```python
for writer in writers:
    content = writer.text_content()
    content = content.strip('')
    content = re.sub(r'\s+', ' ', content)
    print(content)
```

```python
data = [] #숫자바꾸기

for page in range(1, 124):
    res = requests.get(url.format(page))
    root = lxml.html.fromstring(res.text)

    writers = root.cssselect('div.list-dated')

    for writer in writers:
        content = writer.text_content().strip()
        content = re.sub(r'\s+', ' ', content)
        data.append((content))
```

```python
data
```

```python
import pandas as pd
df = pd.DataFrame(data, columns=['writer'])
```

```python
df
```

```python
df.to_csv('기사_2020')
```
